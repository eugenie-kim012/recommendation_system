# 모델에 활용하기 위한 전처리 (실습)

사람이 태어나면 주민등록번호가, 학교에 가면 출석번호가 있듯이 데이터의 관리를 쉽게 하기 위해 번호를 붙여주고 싶습니다. 우리가 다루는 데이터에서는 user와 artist 각각에 번호를 붙이고 싶습니다. 보통 이런 작업을 indexing이라고 합니다. 추천 시스템, 자연어 처리에서 자주 하는 작업들이라 직접 해보시면 좋겠습니다.

(참고) pandas.DataFrame.unique()은 특정 컬럼에 포함된 유니크한 데이터만 모아 줍니다. indexing 작업을 위해 매우 유용합니다.

```python
# 고유한 유저, 아티스트를 찾아내는 코드
user_unique = data['user_id'].unique()        # 중복을 제거한 유저 ID 배열 추출
artist_unique = data['artist'].unique()       # 중복을 제거한 아티스트 이름 배열 추출

# 유저, 아티스트 indexing 하는 코드 (idx는 index의 약자입니다)
user_to_idx = {v: k for k, v in enumerate(user_unique)}     # 유저 이름 → 숫자 인덱스로 매핑하는 딕셔너리 생성
artist_to_idx = {v: k for k, v in enumerate(artist_unique)} # 아티스트 이름 → 숫자 인덱스로 매핑하는 딕셔너리 생성
```

```python
# 인덱싱이 잘 되었는지 확인해 봅니다.
print(user_to_idx['zimin'])    # 358869명의 유저 중 마지막으로 추가된 유저이니 358868이 나와야 합니다.
print(artist_to_idx['maroon5'])
```

```python
# indexing을 통해 데이터 컬럼 내 값을 바꾸는 코드
# dictionary 자료형의 get 함수는 https://wikidocs.net/16 을 참고하세요.

# user_to_idx.get을 통해 user_id 컬럼의 모든 값을 인덱싱한 Series를 구해 봅시다.
# 혹시 정상적으로 인덱싱되지 않은 row가 있다면 인덱스가 NaN이 될 테니 dropna()로 제거합니다.
temp_user_data = data['user_id'].map(user_to_idx.get).dropna()
if len(temp_user_data) == len(data):   # 모든 row가 정상적으로 인덱싱되었다면
    print('user_id column indexing OK!!')
    data['user_id'] = temp_user_data   # data['user_id']을 인덱싱된 Series로 교체해 줍니다.
else:
    print('user_id column indexing Fail!!')

# artist_to_idx을 통해 artist 컬럼도 동일한 방식으로 인덱싱해 줍니다.
temp_artist_data = data['artist'].map(artist_to_idx.get).dropna()
if len(temp_artist_data) == len(data):
    print('artist column indexing OK!!')
    data['artist'] = temp_artist_data
else:
    print('artist column indexing Fail!!')

data
```

# 사용자의 명시적/암묵적 평가

 만약 우리 데이터가 좋아요나 별점처럼 선호도를 명시적(explicit)으로 나타내는 것이라면 참 좋겠지만, 지금 우리가 가지고 있는 데이터는 사용자가 아티스트의 곡을 몇 번 플레이했나 하는 것. 이렇게 서비스를 사용하면서 자연스럽게 발생하는 암묵적(implicit)인 피드백도 사용자의 아이템에 대한 평가를 알 수 있는 단서가 될 수 있음.

```python
# 1회만 play한 데이터의 비율을 보는 코드
only_one = data[data['play']==1]
one, all_data = len(only_one), len(data)
print(f'{one},{all_data}')
print(f'Ratio of only_one over all data is {one/all_data:.2%}')  # f-format에 대한 설명은 https://bit.ly/2DTLqYU
```

```python
# CSR(Compressed Sparse Row) Matrix 그래프 만들기 실습
from scipy.sparse import csr_matrix

num_user = data['user_id'].nunique()
num_artist = data['artist'].nunique()

csr_data = csr_matrix((data.play, (data.user_id, data.artist)), shape= (num_user, num_artist))
csr_data
```

```python
from implicit.als import AlternatingLeastSquares
import os
import numpy as np

# implicit 라이브러리에서 권장하고 있는 부분입니다. 학습 내용과는 무관합니다.
os.environ['OPENBLAS_NUM_THREADS']='1'
os.environ['KMP_DUPLICATE_LIB_OK']='True'
os.environ['MKL_NUM_THREADS']='1'
```

AlternatingLeastSquares 클래스의 __init__ 파라미터를 살펴보겠습니다.

factors : 유저와 아이템의 벡터를 몇 차원으로 할 것인지
regularization : 과적합을 방지하기 위해 정규화 값을 얼마나 사용할 것인지
use_gpu : GPU를 사용할 것인지
iterations : epochs와 같은 의미입니다. 데이터를 몇 번 반복해서 학습할 것인지
1과 4를 늘릴수록 학습 데이터를 잘 학습하게 되지만 과적합의 우려가 있으니 좋은 값을 찾아야 합니다.

```python
# Implicit AlternatingLeastSquares 모델의 선언
als_model = AlternatingLeastSquares(factors=100, regularization=0.01, use_gpu=False, iterations=15, dtype=np.float32)
```

❓ “als_model.fit(csr_data)는 희소행렬을 모델링한다는 말인가요?”
👉 네, 맞습니다.

✅ 구체적으로 설명하자면:
als_model.fit(csr_data)는 다음을 수행합니다:

🔧 입력된 희소행렬(csr_data)의 사용자-아이템 상호작용 데이터를 기반으로,
사용자와 아이템 각각의 잠재 요인(latent factors)을 학습합니다.

즉,

희소행렬 R을 보고

그것을 저차원의 두 행렬 X(user factors), Y(item factors) 로 분해해서

R ≈ X × Yᵀ 의 관계를 모델링하는 겁니다.

🎯 따라서 이 문장은 옳습니다:
"als_model.fit(csr_data)는 희소행렬의 구조를 모델링하고,
그 구조를 설명할 수 있는 사용자·아이템의 잠재 요인 벡터를 학습한다."

| 질문                      | 답변                                                                      |
| ----------------------- | ----------------------------------------------------------------------- |
| 이 코드가 희소행렬을 생성하나요?      | ❌ 아니요. `csr_data`는 이미 생성된 희소행렬입니다.                                      |
| 이 코드가 희소행렬을 **모델링**하나요? | ✅ 맞습니다. 희소행렬의 패턴을 학습해서 미래 추천에 활용합니다.                                    |
| 모델링 결과는 무엇인가요?          | 사용자와 아이템의 잠재 요인 벡터 (`als_model.user_factors`, `als_model.item_factors`) |


```python
print("user_factors shape:", als_model.user_factors.shape)
print("item_factors shape:", als_model.item_factors.shape)
```

```python
#artist_to_idx 를 뒤집어, index로부터 artist 이름을 얻는 dict를 생성합니다.
idx_to_artist = {v:k for k,v in artist_to_idx.items()}
[idx_to_artist[i[0]] for i in similar_artist]
```

```python
from google.colab import drive
drive.mount('/content/drive')

import nbformat
import os

# 정확한 파일명 및 경로
filename = "D+73_iu_fan_recommendation.ipynb"
notebook_path = "/content/drive/MyDrive/Colab Notebooks"
full_path = os.path.join(notebook_path, filename)

# 노트북 로드 (파일 객체로 열기)
with open(full_path, "r", encoding="utf-8") as f:
    nb = nbformat.read(f, as_version=4)

# widgets 제거 (확실하게 처리)
if "widgets" in nb.metadata:
    print("🔧 'metadata.widgets' 제거 중...")
    nb.metadata.pop("widgets", None)  # 없으면 무시하고 넘어감
    print("✅ 제거 완료")
else:
    print("✅ 이미 'metadata.widgets' 없음")

# 다시 저장
with open(full_path, "w", encoding="utf-8") as f:
    nbformat.write(nb, f)

print(f"📁 노트북 저장 완료: {full_path}")
```
